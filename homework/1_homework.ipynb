```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

# 1-1


```python
X = np.random.rand(5,1) * 20
y = -2.345 * X + 3.093

y = y + np.random.normal(size = y.shape ,scale = 3 )

X=X.reshape(-1,1)
y=y.reshape(-1,1)





lr_model = LinearRegression(normalize=True)
lr_model.fit(X,y)



x_line = np.arange(-2,40,0.1).reshape(-1,1)

y_line = lr_model.predict(x_line)

plt.plot(X[:,0],y,'bo')
plt.plot(x_line[:,0],y_line,'r--')
```

```python
X = np.random.rand(5,1) * 20
y = -2.345 * X+ 3.093

y = y + np.random.normal(size = y.shape ,scale = 3 )

X=X.reshape(-1,1)
y=y.reshape(-1,1)



for i in range(2,17):
    X_column_1 = X[:,0].reshape(-1,1)
    X = np.hstack((X,X_column_1**i))


lr_model = LinearRegression(normalize=True)
lr_model.fit(X,y)



x_line= np.arange(0,20,0.1).reshape(-1,1)

for i in range(2,17):
    X_line_column_1 = x_line[:,0].reshape(-1, 1)
    x_line = np.hstack((x_line, X_line_column_1**i))
    
    
y_line = lr_model.predict(x_line)

plt.plot(X[:,0],y,'bo')
plt.plot(x_line[:,0],y_line,'r--')
```
